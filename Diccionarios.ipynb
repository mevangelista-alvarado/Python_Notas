## Diccionarios




La estructura de diccionario nos permite almacenar diferentes tipos de variables es una misma estructura y acceder a ellas mediante "llaves".






Los diccionarios se caracterizan en Python por el uso de llaves $\{\,\}$ para agrupar las variables contenidas y tener la siguiente estructura `{key: value}`.

Por lo regular:
 * La variable `key` son variables del tipo `string` e `int` y
 * La variable `value` tiene mayor libertad y puede se cualquier variable (`string`, `int`, `float`, `list` o `dict`).

Los diccionarios se ordenan acorde a los valores de sus llaves (`keys`) y son mutables.

Para crear un diccionario vacía simplemente no escribimos elementos:

```
diccionario = {}
```

ó tambien podemos escribir

```
diccionario = dict()
```

Como primer ejemplo pensemos en un directorio telefónico, donde las etiquetas son los nombres de nuestros contactos y a través de cada una de ellas se accede al número telefónico:

```
telefonos = {
  'Gaby': 533233,
  'Marina': 555777,
  'Joaquin': 222345,
  'Esteban': 222345,
}
```

y para acceder al teléfono especificamos el nombre del contacto entre paréntesis cuadrados [ ]

```
print(telefonos['Gaby'])
```

Otra manera de acceder a los valores de un diccionario es mediante el método `get()`

```
gaby_telefono = telefonos.get("Gaby")
gaby_telefono
```

La función `get` permite agregar un valor por defaul, por sí no existe la llave en el diccionario y así evitar un error, tal y como ocurre si intento acceder al valo de la llave mediante la notación de corchetes `[]`  

```
miguel_telefono = telefonos.get("Miguel", "No existe la llave")
miguel_telefono
```

```
print(telefonos['Miguel'])
```

Se puede agregar un valor nuevo a un diccionario de la siguiente manera:

```
telefonos["Miguel"] = 5555555
telefonos
```

Otra forma de actulizar un valor de un diccionario es mediante el método `update` el cual actualiza un elemento que ya está en el diccionario, y si no existe todavía, lo añadirá.

```
telefonos.update({"Pepe":78907890})
telefonos
```

También se pueden eliminar elementos de una diccionario.

```
telefonos.pop("Pepe")
telefonos
```

Para utilizar este método debemos estar seguro que la llave exista para no obtener un error.

Tambien es posible vaciar un dicionaro con la sentencia `clear()`

```
_telefonos = telefonos.copy()
_telefonos.clear()
_telefonos
```

Un segundo ejemplo es utilizar diccionarios para agrupar la información de un usuario:

```
usuario_1 = {
  'nombre': 'Gaby',
  'edad': 32,
  'email': 'gaby@email.com',
  'telefono': 555777,
  'nickname':'gaby32'
}

usuario_2 = {
  'nombre': 'Miguel',
  'edad': 32,
  'email': 'miguel@email.com',
  'telefono': 533233,
  'nickname': 'miguel32',
  'nombre': 'miguel'
}
```

y accedemos a la información de cada usuario:

```
print(usuario_1['email'])
print(usuario_2['telefono'])
```

Podemos saber cuales son todas las llaves que contiene el diccionario con la función `keys()`

```
usuario_1.keys()
```

De mismo modo, se pueden obtener todas los valores del diccionario mediante la sentencia `items()`

```
usuario_1.items()
```

_OBS._

Notemos que si un diccionario encuentra una llave duplicada sobre escribirá el valor como ocurre en el siguiente ejemplo.



```
print(usuario_2['nombre'])
```

Por tal razón, debemos tener cuidado de que todas las llaves deben se der únicas.
